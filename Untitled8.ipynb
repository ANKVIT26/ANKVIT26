```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load and Prepare Data
data1 = pd.read_csv('/content/Housing.csv')
data2 = pd.read_csv('/content/House Price India.csv')

# Display first few rows of both datasets
print("Dataset 1:")
print(data1.head())
print("\nDataset 2:")
print(data2.head())

# Feature selection for Dataset 1
features1 = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus']
X1 = data1[features1]
# One-hot encoding for categorical features in Dataset 1
categorical_features = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus']
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for numpy array, handle_unknown to avoid errors with unseen values during prediction
encoded_features = encoder.fit_transform(X1[categorical_features])
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_features))
X1 = X1.drop(categorical_features, axis=1) # Drop original categorical columns
X1 = pd.concat([X1, encoded_df], axis=1) # Concatenate encoded features
y1 = data1['price']  # Adjust column name for target variable if needed

# Feature selection for Dataset 2
features2 = ['living area', 'number of bedrooms', 'number of bathrooms', 'number of floors', 'waterfront present', 'number of views', 'condition of the house', 'grade of the house', 'Area of the house(excluding basement)', 'Area of the basement', 'Built Year', 'Renovation Year', 'Number of schools nearby', 'Distance from the airport']
X2 = data2[features2]
y2 = data2['Price']  # Adjust column name for target variable if needed

# Split data into training and testing sets
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Supervised Learning - Linear Regression
# Initialize and train the model for both datasets
model1 = LinearRegression()
model1.fit(X1_train, y1_train)
model2 = LinearRegression()
model2.fit(X2_train, y2_train)

# Make predictions
y1_pred = model1.predict(X1_test)
y2_pred = model2.predict(X2_test)

# Evaluate the models
mse1 = mean_squared_error(y1_test, y1_pred)
mse2 = mean_squared_error(y2_test, y2_pred)
print(f'Dataset 1 - Mean Squared Error: {mse1}')
print(f'Dataset 2 - Mean Squared Error: {mse2}')

# Plot predictions vs actual values for both datasets
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y1_test, y=y1_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Dataset 1: Actual vs Predicted Prices')

plt.subplot(1, 2, 2)
sns.scatterplot(x=y2_test, y=y2_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Dataset 2: Actual vs Predicted Prices')

plt.tight_layout()
plt.show()

# Logistic Regression (for classification tasks)
# Assuming 'Price' is converted to a binary classification problem
y1_class = (data1['price'] > data1['price'].median()).astype(int)
y2_class = (data2['Price'] > data2['Price'].median()).astype(int)

# Split data into training and testing sets for classification
X1_train, X1_test, y1_train_class, y1_test_class = train_test_split(X1, y1_class, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train_class, y2_test_class = train_test_split(X2, y2_class, test_size=0.2, random_state=42)

# Initialize and train the logistic regression model for both datasets
log_model1 = LogisticRegression()
log_model1.fit(X1_train, y1_train_class)
log_model2 = LogisticRegression()
log_model2.fit(X2_train, y2_train_class)

# Make predictions
y1_class_pred = log_model1.predict(X1_test)
y2_class_pred = log_model2.predict(X2_test)

# Evaluate the models
accuracy1 = accuracy_score(y1_test_class, y1_class_pred)
accuracy2 = accuracy_score(y2_test_class, y2_class_pred)
print(f'Dataset 1 - Accuracy: {accuracy1}')
print(f'Dataset 2 - Accuracy: {accuracy2}')

# Plot predictions vs actual values for both datasets (classification)
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y1_test_class, y=y1_class_pred)
plt.xlabel('Actual Classes')
plt.ylabel('Predicted Classes')
plt.title('Dataset 1: Actual vs Predicted Classes')

plt.subplot(1, 2, 2)
sns.scatterplot(x=y2_test_class, y=y2_class_pred)
plt.xlabel('Actual Classes')
plt.ylabel('Predicted Classes')
plt.title('Dataset 2: Actual vs Predicted Classes')

plt.tight_layout()
plt.show()
```
